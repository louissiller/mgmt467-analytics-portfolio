### Q1 ‚Äî Exploratory vs. Explanatory
**Prompt:** *"Explain the difference between exploratory and explanatory visualization."*
**Hints:** Consider audience, purpose, level of polish, and how much iteration you expect.

üìù **Your response:**
1. Audience:
* Exploratory: The primary audience is the data analyst or the data science team. You're visualizing for yourself to gain understanding.
* Explanatory: The audience is external to the analysis process ‚Äì stakeholders, clients, managers, etc. You're visualizing to communicate findings to others.
2. Purpose:
* Exploratory: To explore the data, identify patterns, discover insights, test hypotheses, and find the story within the data. It's about discovery.
* Explanatory: To explain a specific insight, tell a clear story with the data, support a conclusion, or persuade an audience. It's about communication.
3. Level of Polish:
* Exploratory: Often unpolished, rough, and quick to create. Labels might be missing, colors might not be finalized, and the focus is on functionality rather than aesthetics.
* Explanatory: Highly polished, clean, and professional. Attention is paid to details like titles, labels, legends, color schemes, and overall visual appeal to ensure clarity and impact for the audience.
4. How much iteration you expect:
* Exploratory: High iteration. You'll likely create many different visualizations, trying various chart types and data subsets to see what emerges. It's a process of trial and error.
* Explanatory: Low iteration (for the final version presented to the audience). The iteration happens during the exploratory phase. The final explanatory visualization should be clear and require little to no explanation or interaction from the audience to understand the main point

### Q2 ‚Äî Chart Types & When to Use Them
**Prompt:** *"Describe 5 common chart types (e.g., bar, line, scatter, histogram, map) and the type of data analysis question each is best suited to answer."*

For each chart, list: **Question archetype**, **Data shape**, **When to use it**, **How to use it**, **Pitfalls**, **Alternatives**.

üìù **Your response (template):**
1. **Bar:**
   - Question archetype: **Comparison** (comparing values across different categories) or **Ranking** (showing the order of values).
   - Data shape: Typically one categorical variable and one numerical variable.
   - When to use it: To show the relative amounts of different categories, compare data across discrete groups, or track changes over time for a few categories.
   - How to use it: Plot categories on one axis and the numerical values on the other. Ensure bars have consistent width and spacing. Sort bars if ranking is important.
   - Pitfalls: Can be misleading with too many categories or if the y-axis doesn't start at zero.
   - Alternatives: Column chart (vertical bars), stacked bar chart, dot plot.
2. **Line:**
   - Question archetype: **Trend over Time** (showing how a value changes continuously over a period).
   - Data shape: Typically one time-series variable and one numerical variable.
   - When to use it: To display trends in data over a continuous period, compare trends of different groups over time, or show acceleration/deceleration.
   - How to use it: Plot time on the horizontal axis and the numerical value on the vertical axis. Connect data points with a line. Use multiple lines to compare trends of different categories.
   - Pitfalls: Can be misleading with irregular time intervals or if multiple lines are too close together. Choosing the wrong axis can be misleading and lead to miscommunication. Also wrong scaling can lead to misinterpretation.
   - Alternatives: Area chart, scatter plot with lines (for non-time series data).
3. **Scatter:**
   - Question archetype: **Relationship/Correlation** (showing the relationship between two numerical variables).
   - Data shape: Two numerical variables.
   - When to use it: To identify potential relationships or correlations between two numerical variables, spot outliers, or see the distribution of data points.
   - How to use it: Plot one numerical variable on the x-axis and the other on the y-axis. Each data point represents an observation. Can use color, size, or shape to represent a third variable.
   - Pitfalls: Can be difficult to interpret with too many data points (overplotting) or if there's no clear pattern.
   - Alternatives: Bubble chart (adds a third numerical variable), hexbin plot (for dense data).
4. **Histogram:**
   - Question archetype: **Distribution** (showing the frequency of values within bins for a single numerical variable).
   - Data shape: One numerical variable.
   - When to use it: To understand the shape of the distribution of a single numerical variable, identify peaks and valleys, and see the spread and central tendency of the data.
   - How to use it: Divide the range of the numerical variable into bins (intervals) and draw bars whose height represents the number of data points falling into each bin.
   - Pitfalls: The choice of bin size can significantly impact the appearance of the distribution.
   - Alternatives: Density plot, box plot.
5. **Map:**
   - Question archetype: **Geographic Patterns** (showing how values or events are distributed across a geographic area).
   - Data shape: Geographic data (latitude/longitude, region names) and one or more numerical/categorical variables.
   - When to use it: To visualize data that has a geographic component, show spatial relationships, or highlight regional differences.
   - How to use it: Use geographic boundaries (like states or countries) and color or size to represent the numerical or categorical variable of interest within those areas. Or use points at specific locations with size or color indicating a value.
   - Pitfalls: Can be misleading if not normalized by population or area, or if color scales are not chosen carefully.
   - Alternatives: Choropleth map, symbol map, flow map.

### Data Normalization
Data normalization is a technique used to scale numerical data to a standard range. It involves transforming the values of different features in your dataset to a similar scale, without distorting the differences in the ranges of values or losing information.
You should consider normalizing data before starting several types of analysis and visualization because:
* It helps with algorithms sensitive to scale: Many machine learning algorithms (like k-nearest neighbors, support vector machines, and those using gradient descent) perform better when numerical input variables are on a similar scale. Without normalization, features with larger values might dominate the distance calculations or gradient updates.
* It can improve visualization: When features are on vastly different scales, it can be hard to visualize their relationships effectively on the same plot. Normalization brings them into a comparable range.
* It aids in comparing features: Normalization allows you to compare the relative importance or impact of different features even if their original units or ranges were different.

Common normalization techniques include Min-Max scaling (scaling to a specific range, usually 0 to 1) and Standardization (scaling to have a mean of 0 and a standard deviation of 1). The best approach depends on the specific data and the analysis you plan to perform.

### Create a Synthetic Airline Dataset (if you don't have one)
Use this dataset to prototype visualizations and dashboards. It contains realistic columns for the KPIs you defined. You can always swap in your BigQuery data later.


````python
# Generate a synthetic airline dataset
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

np.random.seed(467)

dates = pd.date_range(start='2024-01-01', end='2024-06-30', freq='D')
carriers = ['AA', 'DL', 'UA', 'WN', 'AS']
origins = ['ATL','DFW','DEN','ORD','LAX','SEA','SFO','PHX','CLT','MCO']
dests =   ['JFK','LGA','BOS','SFO','LAX','SEA','MIA','IAH','IAD','PHL']

rows = []
for d in dates:
    for carrier in carriers:
        # number of flights per day per carrier
        n = np.random.randint(80, 140)
        for _ in range(n):
            o = np.random.choice(origins)
            de = np.random.choice(dests)
            if o == de:
                continue
            distance = np.random.randint(300, 2800)
            seats = np.random.choice([90,120,150,180,210])
            load_factor = np.clip(np.random.normal(0.82, 0.07), 0.55, 0.98)
            pax = int(seats * load_factor)
            dep_delay = int(np.random.normal(8, 15))
            arr_delay = dep_delay + int(np.random.normal(0, 12))
            dep_delay = max(dep_delay, -5)
            arr_delay = max(arr_delay, -10)
            cancelled = np.random.rand() < 0.025
            diverted = np.random.rand() < 0.005
            revenue = pax * np.random.uniform(120, 420)
            casm = np.random.uniform(0.08, 0.16)  # cost per available seat mile
            cost = casm * seats * distance
            fuel_burn = distance * np.random.uniform(3.0, 4.5)  # gallons per seat-mile proxy
            nps = np.random.normal(35, 15)  # if you want a customer metric proxy
            rows.append({
                'date': d,
                'carrier': carrier,
                'origin': o,
                'dest': de,
                'distance_miles': distance,
                'seats': seats,
                'passengers': pax,
                'dep_delay_min': dep_delay,
                'arr_delay_min': arr_delay,
                'cancelled': int(cancelled),
                'diverted': int(diverted),
                'revenue_usd': round(revenue, 2),
                'cost_usd': round(cost, 2),
                'fuel_burn_gal': round(fuel_burn, 1),
                'nps': round(nps, 1)
            })

df_flights = pd.DataFrame(rows)

# Derived KPIs
df_flights['on_time'] = (df_flights['arr_delay_min'] <= 15).astype(int)
df_flights['load_factor'] = (df_flights['passengers'] / df_flights['seats']).round(3)
df_flights['yield_usd_per_pax_mile'] = (df_flights['revenue_usd'] / (df_flights['passengers'] * df_flights['distance_miles']).clip(lower=1)).round(4)
df_flights['asm'] = df_flights['seats'] * df_flights['distance_miles']  # available seat miles
df_flights['casm'] = (df_flights['cost_usd'] / df_flights['asm']).round(4)
df_flights['month'] = df_flights['date'].dt.to_period('M').astype(str)

df_flights.head()
````

````python
# Calculate On-Time Rate by carrier
on_time_rate_by_carrier = df_flights.groupby('carrier')['on_time'].mean().sort_values(ascending=False) * 100

plt.figure(figsize=(10, 6))
sns.barplot(x=on_time_rate_by_carrier.index, y=on_time_rate_by_carrier.values, palette='viridis')

plt.title('On-Time Rate by Carrier', fontsize=16)
plt.xlabel('Carrier', fontsize=12)
plt.ylabel('On-Time Rate (%)', fontsize=12)
plt.ylim(0, 100)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Add percentage labels on bars
for index, value in enumerate(on_time_rate_by_carrier.values):
    plt.text(index, value + 1, f'{value:.1f}%', ha='center', va='bottom', fontsize=10)

plt.show()
````

````python
# Aggregate to daily average arrival delay
daily_avg_delay = df_flights.groupby('date')['arr_delay_min'].mean()

# Calculate 7-day rolling average
rolling_avg_delay = daily_avg_delay.rolling(window=7).mean()

plt.figure(figsize=(12, 7))
plt.plot(daily_avg_delay.index, daily_avg_delay.values, label='Daily Average Delay', alpha=0.7)
plt.plot(rolling_avg_delay.index, rolling_avg_delay.values, label='7-Day Rolling Average', color='orange')

plt.title('Daily Average Arrival Delay with 7-Day Rolling Average', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Average Arrival Delay (minutes)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show()
````

````python
plt.figure(figsize=(10, 6))
sns.histplot(df_flights['arr_delay_min'], bins=50, kde=True)

plt.title('Distribution of Arrival Delays', fontsize=16)
plt.xlabel('Arrival Delay (minutes)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.axvline(0, color='red', linestyle='--', label='On-Time')

# Calculate percent of flights within [-5, 15] minutes
on_time_flights = df_flights[(df_flights['arr_delay_min'] >= -5) & (df_flights['arr_delay_min'] <= 15)]
percent_on_time = (len(on_time_flights) / len(df_flights)) * 100

plt.text(18, plt.gca().get_ylim()[1] * 0.8, f'{percent_on_time:.1f}% within [-5, 15] min', color='green', fontsize=12)
plt.legend()
plt.tight_layout()
plt.show()
````

````python
# Compute mean load factor by route and take the top 15
route_load_factor = df_flights.groupby(['origin', 'dest'])['load_factor'].mean().reset_index()
route_load_factor['route'] = route_load_factor['origin'] + '-' + route_load_factor['dest']
top_15_routes = route_load_factor.nlargest(15, 'load_factor')

plt.figure(figsize=(12, 8))
sns.barplot(x='load_factor', y='route', data=top_15_routes.sort_values(by='load_factor', ascending=False), palette='viridis')

plt.title('Top 15 Routes by Average Load Factor', fontsize=16)
plt.xlabel('Average Load Factor', fontsize=12)
plt.ylabel('Route', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
````

````python
plt.figure(figsize=(10, 8))
scatter = sns.scatterplot(data=df_flights, x='cost_usd', y='revenue_usd', hue='casm', palette='viridis', s=50, alpha=0.6)

plt.title('Cost vs. Revenue (Colored by CASM)', fontsize=16)
plt.xlabel('Cost (USD)', fontsize=12)
plt.ylabel('Revenue (USD)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)

# Add a y=x reference line
max_val = max(df_flights['cost_usd'].max(), df_flights['revenue_usd'].max())
plt.plot([0, max_val], [0, max_val], color='red', linestyle='--', label='Cost = Revenue')

plt.legend(title='CASM')
plt.tight_layout()
plt.show()
````
